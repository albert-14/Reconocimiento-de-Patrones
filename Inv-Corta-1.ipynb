## Aggregation


A continuación, se utilizarán una serie de funciones que nos ayudarán a visualizar características de nuestro set de datos.

### Conteo

Esta función realiza el conteo de los valores en cada columna.

```python
data.count()
```

De igual manera, podemos realizar el conteo por únicamente una columna.

```python
data.Age.count()
```

### Suma

Esta función suma los valores de cada columna o de una columna en específico tal y como se muestra a continuación.

```python
data.Age.sum()
```

### Máximo y Mínimo

Estas funciones identifican el valor más grande y el más pequeño de cada columna o de una en específico.

```python
data.Age.max()
```

```python
data.Age.min()
```

### Media y Mediana

Estas funciones calculan valores estadísticos como lo son la media y la mediana.
Se puede calcular para todas las columnas o para alguna en específico.

```python
data.Age.mean()
```

```python
data.Age.median()
```

## Grouping

Estas funciones son muy útiles en el análisis de datos ya que permiten segmentar los datos por categorías o grupos de interés.
A continuación, se muestra un ejemplo de agrupamiento por edad y el cálculo de la media.

```python
group1 = data.groupby('Age').mean()[['Annual Income (k$)']]

group2 = data.groupby('Age').mean()[['Spending Score (1-100)']]

group2
```

# Análisis Exploratorio de Datos

Con el objetivo de analizar los datos de manera gráfica, se utilizará matplotlib.

Tal y como se estudió en clase, el análisis exploratorio de datos representa el uso de métodos cuantitativos y visuales para comprender mejor los datos sin tener que asumir cosas.

Esto se puede llevar a cabo de alguna de estas tres maneras:

1) Visualización de un resumen estadístico

2) Exploración visual de relaciones para detectar atributos

3) Mediante diagramas de dispersión para observar cualquier agrupamiento

A continuación, se graficará la relación entre la edad de los clientes versus el promedio de ingreso anual.

Es interesante observar como los clientes con edades de entre 25 años y 35 años representan los ingresos más altos.

```python
data.groupby('Age').mean()[['Annual Income (k$)']]
plt.bar(data['Age'],data['Annual Income (k$)'])
plt.ylabel('Annual Income [$]')
plt.xlabel('Customer Age')
plt.title('Customer Age vs Annual Income')
```

Posteriormente, se grafica el comportamiento entre las edades de los clientes versus el promedio de calificación para el gasto de los cliente.

```python
data.groupby('Age').mean()[['Spending Score (1-100)']]
plt.bar(data['Age'],data['Spending Score (1-100)'])
plt.ylabel('Spending Score [1-100]')
plt.xlabel('Customer Age')
plt.title('Customer Age vs Spending Score')
```
